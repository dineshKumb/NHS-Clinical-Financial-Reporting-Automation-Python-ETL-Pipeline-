## Copy all financial files into respective person's Bronze folder

```python
import os
import shutil
import pandas as pd
import Config

# ----------------------------
# CONFIG
# ----------------------------
FINANCE_DIR = r'{}\finance'.format(Config.BASE_PATH) # folder containing 325 finance files
PERSON_FOLDERS_DIR = r'{}\Working environment'.format(Config.BASE_PATH)  # base folder with Person_X/Bronze
LOG_FILE = r'{}\area_assignment_log.csv'.format(Config.BASE_PATH)

# ----------------------------
# Load the assignment log
# ----------------------------
log_df = pd.read_csv(LOG_FILE)

# ----------------------------
# Copy files into Bronze folder
# ----------------------------
for idx, row in log_df.iterrows():
    area_file = row['area_file']
    person_id = row['person_id']

    # Source file path
    src_file = os.path.join(FINANCE_DIR, area_file)
    
    # Destination Bronze folder path
    dest_folder = os.path.join(PERSON_FOLDERS_DIR, person_id, "Bronze")
    os.makedirs(dest_folder, exist_ok=True)  # just in case
    
    # Destination file path
    dest_file = os.path.join(dest_folder, area_file)
    
    # Copy file
    shutil.copy2(src_file, dest_file)

print("All assigned finance files have been copied to respective Bronze folders.")
```

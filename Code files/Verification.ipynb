## verify if all required paths are present or not.

```python
import os
import Config
# ----------------------------
# CONFIG
# ----------------------------

BASE_PATH = r'{}\Working environment'.format(Config.BASE_PATH)
NUM_PERSONS = 36
SUBFOLDERS = ["Bronze", "Silver", "Gold"]

# ----------------------------
# Create/verify folders
# ----------------------------
for person_id in range(1, NUM_PERSONS + 1):
    person_folder = os.path.join(BASE_PATH, f"Person_{person_id}")
    
    # Create main person folder if not exists
    if not os.path.exists(person_folder):
        os.makedirs(person_folder)
        print(f"Created folder: {person_folder}")
    else:
        print(f"Folder already exists: {person_folder}")
    
    # Create/verify subfolders
    for subfolder in SUBFOLDERS:
        subfolder_path = os.path.join(person_folder, subfolder)
        if not os.path.exists(subfolder_path):
            os.makedirs(subfolder_path)
            print(f"  Created subfolder: {subfolder_path}")
        else:
            print(f"  Subfolder already exists: {subfolder_path}")

print("\nAll person folders and subfolders verified/created successfully.")
```
